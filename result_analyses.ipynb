```python
with open(f'kde_5_beatit.npy', 'rb') as file:
    integrals = np.load(file)

period = 50
logs_integrals = np.log2(integrals)
logs_integrals = 1 / integrals

wsize = 64
w = signal.windows.hann(wsize)
means = np.convolve(logs_integrals, w, mode='valid')
indices = np.arange(len(logs_integrals))
ma_indices = indices[:len(means)] + (period - 1) / 2



plt.title('Beat It - Michael Jackson\nEntropy')
plt.figsize=(19, 10)
plt.plot(ma_indices, means, label=f'Curva de Entropia', marker='None', color='indigo', linewidth=1)
plt.vlines(beatit_annotations,min(means), max(means), label='Mudanças', colors='r', linestyles='dashed',  linewidth=.7)
plt.legend()
```
